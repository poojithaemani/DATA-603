```python
import pyspark
import requests
import re

def display_URL(url):
    html_img = requests.get(url).text
    images = re.findall(r'src=["\'](http[s]?://.*?)(?=["\'])', html_img)
    rdd = sc.parallelize(images)
    counts = rdd.map(lambda word: (word, 1)).reduceByKey(lambda x, y: x + y)
    image_counts = counts.collect()

    print(f'There are {len(images)} images at the {url} site')
    print('The images are:')
    for img in image_counts:
        print(img)

# Initialize SparkContext
sc = pyspark.SparkContext()

# List of URLs to analyze
urls = ['https://www.yahoo.com/', 'https://www.umbc.edu/', 'https://www.netflix.com']

# Display information for each URL
for url in urls:
    display_URL(url)

# Stop SparkContext
sc.stop()
```
